```python
Select(driver.find_element(By.ID, 'CPHPage_ddFinyear')).select_by_visible_text('2022-2023')
driver.find_element(By.ID, 'CPHPage_btnShow').click()
dropdown_states = Select(driver.find_element(By.ID, 'CPHPage_ddState'))

# Get all the options in the dropdown and retrieve their text
states = [state.text for state in dropdown_states.options]
states = states[1:]
df_b = pd.DataFrame()
df_d = pd.DataFrame()
```

```python
for state in states:
    print(f"Selecting: {state}")
    
    # Select the option by its visible text
    Select(driver.find_element(By.ID,"CPHPage_ddState")).select_by_visible_text(state)
    sleep(1)
    
    driver.find_element(By.ID, 'CPHPage_btnShow').click()
    sleep(5)

    table_d = pd.read_html(driver.find_element(By.ID, 'tableReportTable').get_attribute("outerHTML"))[0]
    table_d.columns = table_d.columns.get_level_values(1)
    table_d.insert(1, 'State', state)
    df_d = pd.concat([df_d, table_d], axis=0)    
    
    district1 = Select(driver.find_element(By.ID, 'CPHPage_ddDistrict')).options[1].text
    sleep(1)
    Select(driver.find_element(By.ID, 'CPHPage_ddDistrict')).select_by_visible_text(district1)
    sleep(1)
    
    driver.find_element(By.ID, 'CPHPage_btnShow').click()
    sleep(1)
    
    driver.find_element(By.ID, 'CPHPage_chkBlock').click()
    sleep(1)
    
    driver.find_element(By.ID, 'CPHPage_btnShow').click()
    sleep(15)
    
    table_b = pd.read_html(driver.find_element(By.ID, 'tableReportTable').get_attribute("outerHTML"))[0]
    table_b.columns = table_b.columns.get_level_values(1)
    table_b.insert(1, 'State', state)
    df_b = pd.concat([df_b, table_b], axis=0)
```

```python
df_d['District'] = df_d['District'].str.title()
```

```python
df_b['District'] = df_b['District'].str.title()
df_b['Block'] = df_b['Block'].str.title()
```

```python
df_d = df_d[~(df_d['District'] == 'Total')]
df_d = df_d[~(df_d['District'] == 'National Average')]
df_d = df_d[~(df_d['District'] == 'State Average')]
```

```python
df_b = df_b[~(df_b['District'] == 'Total')]
df_b = df_b[~(df_b['District'] == 'National Average')]
df_b = df_b[~(df_b['District'] == 'State Average')]
```
